# 머신런닝이란?

##1. 머신러닝의 정의
기꼐가 일일이 코드로 명시하지 않은 동작을 **데이터**로부터 **학습**하여 실행할 수 있도록 하는 **알고리즘**을 개발하는 연구 분야 (아서 사무엘, 1959)



##2. 머신러닝의 3요소
Machine Learning = Representation + Evaluation + Optimization

* Representation: 모델의 가설 공간 (The model landscape)
* Evaluation: 가설 공간에서의 모델 평가기준 (Preferences over the landscape)
* Optimization: 평가기준에 부합하기 위한 최적화 (Strategy for fulfilling preferences)


---
* 알고리즘: 물제를 풀기 위한 의사 결정 과정
* 데이터: 객관적인 기준을 위한 정답과 요인들
* 학습: 데이터로부터 최적의 값을 찾는 과정
---

## 3. 머신러닝 알고리즘의 종류
* 지도 학습: Supervised Learning: 정답이 있는 문제를 풀 경우
    * 정답의 종류에 따른 지도 학습
    * 회귀: Regression: 정답이연속형변수일때
        * 정답이 연속형 변수일 때
        * 값이 정수처럼 명확하지 않음
        * 예시: 키를 이용해 몸무게를 예측하기
    * 분류: Classification: 정답이 비연속형(범주형) 변수일때
        * 범주형 변수란? 값이 정수처럼 명확함 (예: 성별, 도시 ...)
        * 예시: 키와 몸무게를 이용해 성별을 맞추기

* 비지도 학습: Unsupervised Learning: 정답이 없는 문제를 풀 경우
    * 군집분석: Clustering: 주어진 데이터가 어떻게 구성 되었는지 알아내려는 분석 방법
    * 강화학습: Reinforcement Learning: 행동에 따른 보상을 최대화 시키는 학습 방법

# 머신러닝 소개

## 모델
* 모델의 정의
    * 어떤 X가 주어 졌을 때 f라는 함수를 통해 y라는 값을 도출하는 과정
    * 이 때의 함수 f를 모델 또는 알고리즘이라고 부름

* 모델의 수식
    * y =f(x)
    * x = 데이터
    * y = 예측값

* 모델의 목적: 데이터를 이용해 값을 예측
* 모델의 평가: 모델이 값을 잘 예측하는지 평가



## 데이터의 종류
* Train Data
    * 학습에 사용되는 데이터
* Test Data
    * 학습에 사용되지 않은 데이터
    * 최종 모델의 성능을 검증하기 위한 데이터
    * 모델이 실제로 잘 예측하는지 알기 위해서는 학습에 사용되지 않은 데이터를 이용해 평가
* Valid Data
    * 학습이 완료된 모델을 검증하기 위한 데이터
    * 학습에 사용되지는 않지만 관여하는 데이터
    * Valid 데이터에 Overfitting 될 수도 있음

## 모델의 평가와 데이터의 관계
* Underfitting: 과소적합: 학습 데이터를 잘 맞추지 못하는 현상
* Overfitting: 과대적합: 학습 데이터는 잘 맞추지만 학습데이터 외에는 잘 맞추지 못하는 현상

* Underfitting을 확인하는 방법
    * Train data로 학습된 모델을 Train data로 평가
    * Train data를 잘 맞추지 못 한다면 Underfitting 상태
* Overfitting을 확인하는 방법
    * Train Data을 잘 학습한 모델을 Test data로 평가
    * Train Data는 잘 맞추지만 Test Data를 잘 맞추지 못한다면 Overfitting 상태
* Data split: 데이터를 Train data와 Test data로 나누는 것

## Cross Validation
Valid 데이터를 고정하지 않고 계속해서 변경함으로써 Overfitting  되는 것을 막기 위한 방법

* Cross Validation 종류
    * LOOCV (Leave One Out Cross Validation)
        * 하나의 데이터를 제외하고 모델을 학습한 후 평가
        * 데이터 개수 만큼의 모델을 학습해야 한다
        * 데이터가 많을 경우 시간이 오래 걸린다.
