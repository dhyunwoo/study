```python
#사체실험 환경정보 불러오기
import pandas as pd
env = pd.read_csv('p19env.csv')
#env.columns
```

```python
import matplotlib.pyplot as plt

#linewidth = 5 선 굵기
#color = 'red' 선 색
#marker = 'o' 값 마커
#markersize = 1 마커 크기
#linestyle = ':' 라인 스타일 점선, 등등

plt.plot(env['duration'], env['weight'], 
         linewidth = 5, 
         color = 'g', 
         marker = 'o', 
         markersize = 1, 
         linestyle = ':'
         )
```

```python
#선 없이 스팟만 찍을 때
plt.plot(env['duration'], env['weight'], 'o')
```

```python
#그래프에 제목 넣기
#fontsize = 10 제목 글자 사이즈

plt.plot(env['duration'], env['weight'])
plt.title('weight loss', fontsize = 25)
plt.xlabel('duration (day)', fontsize = 15) #x축 제목
plt.ylabel('weight (kg)', fontsize = 15) #y축 제목

plt.grid(True) #그래프에 그리드 넣기

#그래프에 그리드 넣기
#x축에만 그리드를 넣을 때 plt.grid(True, axis = 'x')
#Y축에만 그리드를 넣을 때 plt.grid(True, axis = 'y')
```

```python
#특정 구간만 그래프를 보기
plt.plot(env['duration'], env['weight'])
plt.xlim(0, 100)
plt.ylim(10, 60)
```

```python
#각 축마다 글꼴, 크기 변경
plt.plot(env['duration'], env['weight'])
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
```

```python
#그래프 위에 텍스트 입력
plt.plot(env['duration'], env['weight'])
plt.text(15, 58, 'loss', fontsize = 10, color = 'r')
```

```python
#matplotlib의 폰트를 Nanum 폰트로 지정
import matplotlib.pyplot as plt

plt.rcParams['font.family']=['NanumBarunGothic', 'sans-serif']
plt.rcParams['axes.unicode_minus'] = False
```

```python
#한글 패치를 해주었기 때문에 한글 출력 가능

plt.plot(env['duration'], env['weight'])
plt.title('시간에 따른 무게감소 그래프', fontsize = 20) 
plt.xlabel('일수 (일)', fontsize = 15) #x축 제목
plt.ylabel('무게 (kg)', fontsize = 15) #y축 제목

plt.grid(True) #그래프에 그리드 넣기

#그래프에 그리드 넣기
#x축에만 그리드를 넣을 때 plt.grid(True, axis = 'x')
#Y축에만 그리드를 넣을 때 plt.grid(True, axis = 'y')
```

```python
#여러데이터를 한 차트에 올리기
#label = '무게' 범례에 들어갈 범레제목, plt.legend를 같이 써줘야함

plt.plot(env['duration'], env['weight'], color = 'b', label = '무게')
plt.plot(env['duration'], env['air_tmp'], color = 'r', label = '기온')
plt.legend('무게', '기온')
```

```python
# 다른 방법의 범례 표시
# loc = 'lower right' 범례의 위치 표시

plt.plot(env['duration'], env['weight'], color = 'b')
plt.plot(env['duration'], env['air_tmp'], color = 'r')
plt.legend(['무게', '기온'], fontsize = 15, loc = 'lower right')
```

```python
#보조축 넣기
#차트 만들기
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
chrt1 = ax1.plot(env['duration'], env['weight'], color = 'b')
chrt2 = ax2.plot(env['duration'], env['air_tmp'], color = 'r')

#보조축 제목 넣기
ax1.set_xlabel('duration') 
ax1.set_ylabel('weight(kg)') 
ax2.set_ylabel('Temp(℃)')

#범례 넣기
chrt = chrt1 +chrt2
ax1.legend(chrt, ['weight', 'airTemp'], loc = 'lower right')

plt.show()
```

```python
#천이표 만들기
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/Colab/study/dataset

import pandas as pd
env = pd.read_csv('p19env.csv')
spm = pd.read_csv('p19spm.csv')
list = pd.read_csv('splist.csv')

a = pd.merge(spm, list, how = 'left', on = 'species_name_ko')
ins = pd.DataFrame(a)

b = pd.merge(env, ins, how = 'left', on = 'date')
suc = pd.DataFrame(b)
```

```python
pivot = pd.pivot_table(suc, 
                       values = 'count', 
                       index = ['species_name_en'], 
                       columns = ['duration'], 
                       fill_value=0,
                       aggfunc = ('sum'))

#히트맵으로 보기
#linewidths=1 #구분선 추가

sns.heatmap(pivot, cmap='Blues')
```
