```python
# 월별 데이터가 3-5월까지 봄, 6-8까지 여름, 9-11월 까지 가을 나머지 값들은 겨울로 지칭하여 season칼럼을 만들어라

# 조건 함수 만들기
def case_function(x): 
    if x < 6:
        return 'spring'
    elif 9> x >5:
        return 'summer' 
    else:
        return 'fall'    
```

```python
ins['season'] = ins['month'].apply(case_function)
```

```python
ins
```

```python
#월 데이터에서숫자를 4월, 5월 이런식으로 출력되게 mm 칼럼을 만들어라: 1대1 매칭으로 변환!
#map 함수 이용

a = {3 : '3월', 4:'4월', 5: '5월', 6: '6월', 7: '7월', 8: '8월', 9: '9월', 10: '10월', 11: '11월', 12: '12월'}
ins['mm'] = ins['month'].map(a)
```
