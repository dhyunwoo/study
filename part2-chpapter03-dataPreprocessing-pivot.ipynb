```python
#colab에서 .csv 파일 가져오기

from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/Colab

import pandas as pd
env = pd.read_csv('p19env.csv')
spm = pd.read_csv('p19spm.csv')
list = pd.read_csv('splist.csv')
```

```python
#먼저 표본데이터 spm에 명집 데이터 list를 조인한다.
#조인된 데이터프레임은 ins로 변수 지정한다.

a = pd.merge(spm, list, how = 'left', on = 'species_name_ko')
ins = pd.DataFrame(a)
```

```python
#학명과 표본데이터가 머지된 ins 데이터프레임과 환경데이터 env를 날짜 date를 기준으로 머지한다.
#조인된 최종 데이터프레임은 suc로 변수 지정한다.

b = pd.merge(env, ins, how = 'left', on = 'date')
suc = pd.DataFrame(b)

#조인된 데이터만 필요할 때 csv파일로 생성
#suc.to_csv('succession-rawdata.csv')
```

```python
#피벗테이블 만들기
#세로축은 학명-종명, 가로축은 기간이 표시되게 한다.
#fill_value=0 비어 있는 값은 0으로 표시한다.

pivot = pd.pivot_table(suc, 
                       values = 'count', 
                       index = ['No','Order_name_En', 'Order_name_ko', 
                                'Family_name_En', 'Family_name_Ko', 
                                'Genus_name_en',
                                'species_name_en', 'species_name_ko'], 
                       columns = ['duration'], 
                       fill_value=0,
                       aggfunc = ('sum'))
```
